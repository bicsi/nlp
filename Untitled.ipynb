```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
COL_NAMES = ["idx", "text", "l", "r", "sub", "n1", "n2", "c1", "c2", "p"]
train_df = pd.read_csv('data/train_full.txt', sep='\t', 
                       header=None, names=COL_NAMES)
```

```python
train_df["sub_token_len"] = [len(s.split()) for s in train_df["sub"]]
train_df["sub_char_len"] = [len(s) for s in train_df["sub"]]
```

```python
all_words = [w for sub in train_df["sub"] for w in sub.split()]
token_counts = pd.Series(all_words).value_counts()
print(len(token_counts))
print(token_counts)
```

```python
small_subs = train_df["sub_char_len"] < 20

plt.title("Probability vs sub length")
plt.hist2d(
    train_df[small_subs]["sub_char_len"], 
    train_df[small_subs]["p"], 
    bins=10,
    norm=mpl.colors.LogNorm(),
)
plt.show()
```

```python
plt.title("Probability vs sub token length")
plt.hist2d(
    train_df["sub_token_len"] - 1, 
    train_df["p"], 
    bins=[10, 10],
    norm=mpl.colors.LogNorm(),
)
plt.show()
```
